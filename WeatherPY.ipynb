# WeatherPy

```python
# Import dependencies
import matplotlib.pyplot as plt
import openweathermapy as ow
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
#cities = cities[0:60]
len(cities)
```

### Perform API Calls

```python
# Define url and units to use in query url
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

# Define settings and query url to use for requests
settings = {"units": "imperial", "appid": api_key}
query_url = f"{url}appid={api_key}&units={units}&q="

# Retrieve data from API using a loop of each city
weather = []
#weather_set = []

print("Beginning Data Retrieval\n--------------------------")
i = 1

for x in range(0, len(cities)):
    
    if x % 30 == 0:
        time.sleep(5)
    
    # Use try/except to print "City not found" for any errors in retrieval
    try:
        
        # Make request for data
        current_weather = ow.get_current(cities[x], **settings)
        
        # Identify data needed and add to empty set
        summary = ["name", "clouds.all", "sys.country", "dt", "main.humidity", "coord.lon", "coord.lat", "main.temp_max", "wind.speed"]
        weather.append(current_weather(*summary))
        
        # Calculate Record & Set numbers to print and display
        n = len(weather)
            
        if n % 50 == 0:
            z = 49
            i = int(n/51) + 1

        else:
            z = n - (int(n/50)*50)- 1
            i = int(n/50) + 1
   
        print(f"Processing Record {z} for Set {i} | {cities[x]}")
             
    except:
        print("City not found. Skipping...")
        
print("--------------------------\nData Retrieval Complete\n--------------------------")
```

### Convert Raw Data to Dataframe

```python
# Convert retrieved data into dataframe and rename columns
weather_df = pd.DataFrame(weather)
ids = ["City", "Cloudiness", "Country", "Date", "Humidity", "Long", "Lat", "Max Temp(°F)", "Wind Speed"]
for n in range(0, len(list(weather_df))):
    weather_df = weather_df.rename(columns={n:ids[n]})
    
# Create csv from city data
weather_df.to_csv('WeatherData.csv', index=False)

# Display dataframe to verify
weather_df.head()
```

```python
# Find count of cities to verify if enough found
weather_df.count()
```

```python
# Convert timestamp to date
weather_df["Date"] = pd.to_datetime(weather_df["Date"], unit='s')
weather_df["Date"] = weather_df["Date"].apply(lambda x: x.date())
weather_df.head()
```

### Plotting the Data

#### Latitude vs. Temperature Plot

```python
# Identify data to use for x and y axis
x_axis = weather_df["Lat"]
y_axis = weather_df["Max Temp(°F)"]
dt = weather_df["Date"][0]

# Plot graph using information identified above and format
with plt.rc_context({'axes.edgecolor':'white', 'xtick.color':'dimgray', 'ytick.color':'dimgray'}):
    fig, ax = plt.subplots()
#ax.set_facecolor('ghostwhite')
ax.scatter(x_axis, y_axis, color = 'steelblue', edgecolors = 'darkslategrey', alpha = .85)
plt.title(f"City Latitude Vs Max Temperature ({dt})", color = 'dimgray')
ax.set_xlabel("Latitude", color = 'dimgray')
ax.tick_params( colors='dimgray')
ax.set_ylabel("Maximum Temperature (F)", color = 'dimgray')
plt.grid(color = 'gainsboro')
ax.set_axisbelow(True)

# Save figure and print
plt.savefig("../Images/LatitudeVsMaxTemp")
plt.show()
```
